```python
import pandas as pd

df = pd.read_csv("../backend/csv/stations_infos.csv", low_memory=False)

df
```

```python
df_unique = df.drop_duplicates(subset=["nom", "lat", "lon"])

df_unique
```

```python
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt

geometry = [Point(xy) for xy in zip(df_unique["lon"], df_unique["lat"])]
gdf = gpd.GeoDataFrame(df_unique, geometry=geometry, crs="EPSG:4326")
world = gpd.read_file("../data/geojson/world.geojson")
france = world[world.name == "France"]

fig, ax = plt.subplots(figsize=(10, 10))
france.plot(ax=ax, color='lightgrey')
gdf.plot(ax=ax, color='red', markersize=10)


plt.title("Visualisation des stations en France")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
df["date"] = pd.to_datetime(df["DATE"].astype(str), format="%Y%m")
df["nom"] = df["nom"].astype(str)
df = df.dropna(subset=["TN", "TX", "TM"])
df["TN"] = df["TN"].str.replace(",", ".")
df.loc[:, "TN"] = df["TN"].astype(float).copy()
df["TX"] = df["TX"].str.replace(",", ".")
df.loc[:, "TX"] = df["TX"].astype(float).copy()
df["TM"] = df["TM"].str.replace(",", ".")
df.loc[:, "TM"] = df["TM"].astype(float).copy()
df
```

```python
plt.figure(figsize=(14, 8))

df.loc[:, "year"] = df["date"].dt.year
df_filter = df[df["year"] <= 2024]

df_moyenne = df_filter.groupby("year")["TN"].mean().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(df_moyenne["year"], df_moyenne["TN"], marker='o', linestyle='-', color='steelblue')
plt.title("Température minimale moyenne annuelle (TN) - Toutes stations confondues")
plt.xlabel("Année")
plt.ylabel("Température minimale moyenne (°C)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_filter.loc[:, "TX"] = df_filter["TX"].astype(float).copy()
df_moyenne = df_filter.groupby("year")["TX"].mean().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(df_moyenne["year"], df_moyenne["TX"], marker='o', linestyle='-', color='steelblue')
plt.title("Température maxmimum moyenne annuelle (TX) - Toutes stations confondues")
plt.xlabel("Année")
plt.ylabel("Température maxmimum moyenne (°C)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_max_tx = df_filter.groupby("year")["TX"].max().reset_index()
df_min_tx = df_filter.groupby("year")["TX"].min().reset_index()

df_min_tn = df_filter.groupby("year")["TN"].max().reset_index()
df_max_tn = df_filter.groupby("year")["TN"].min().reset_index()


fig, ax = plt.subplots(2, 1, figsize=(12, 10))

ax[0].plot(df_max_tx["year"], df_max_tx["TX"], label="Max TX", color="red", linestyle='-', marker='o')
ax[0].plot(df_min_tx["year"], df_min_tx["TX"], label="Min TX", color="blue", linestyle='-', marker='o')
ax[0].set_title("Température Maximale (TX) : Min et Max par année")
ax[0].set_xlabel("Année")
ax[0].set_ylabel("Température Maximale (°C)")
ax[0].grid(True)
ax[0].legend()


ax[1].plot(df_max_tn["year"], df_max_tn["TN"], label="Max TN", color="red", linestyle='-', marker='o')
ax[1].plot(df_min_tn["year"], df_min_tn["TN"], label="Min TN", color="blue", linestyle='-', marker='o')
ax[1].set_title("Température Minimale (TN) : Min et Max par année")
ax[1].set_xlabel("Année")
ax[1].set_ylabel("Température Minimale (°C)")
ax[1].grid(True)
ax[1].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np

stations_random = df["nom"].dropna().unique()
stations_sampled = np.random.choice(stations_random, 5, replace=False)

df_sampled = df[df["nom"].isin(stations_sampled)]
df_sampled = df_sampled[df_sampled["year"] <= 2024]

df_aggregated = df_sampled.groupby(["year", "nom"]).agg({
    "TN": "mean",
    "TX": "mean",
    "TM": "mean"
}).reset_index()

fig, ax = plt.subplots(figsize=(12, 6))
station_record = None
for station in stations_sampled:
    station_record = station
    station_data = df_aggregated[df_aggregated["nom"] == station]

    # Tracer les courbes pour chaque station
    ax.plot(station_data["year"], station_data["TN"], label=f"TN - {station}", alpha=0.7)
    ax.plot(station_data["year"], station_data["TX"], label=f"TX - {station}", alpha=0.7)
    ax.plot(station_data["year"], station_data["TM"], label=f"TM - {station}", alpha=0.7)
    break

ax.set_title(f"Moyennes annuelles de TN (temp. minimum), TX (temp. maximum) et TM (temp. moyenne) pour la station {station_record} (1950 à aujourd'hui)")
ax.set_xlabel("Année")
ax.set_ylabel("Températures (°C)")
ax.legend(loc="best", bbox_to_anchor=(1, 1), fontsize=10)
ax.grid(True)

# Ajuster l'affichage
plt.tight_layout()
plt.show()
```

## Conclusion

La visualisation par station au niveau de la moyenne annuelle de température minimum, maximum et moyenne est la plus intéressante pour voir une tendance de l'augmentation de la température au fil des années.
