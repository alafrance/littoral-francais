```python
import folium
import numpy as np

LON_1 = np.array(ds['LON_1'].values)
LAT_1 = np.array(ds['LAT_1'].values)
LON_2 = np.array(ds['LON_2'].values)
LAT_2 = np.array(ds['LAT_2'].values)
LON_3 = np.array(ds['LON_3'].values)
LAT_3 = np.array(ds['LAT_3'].values)

centres_lon = (LON_1 + LON_2 + LON_3) / 3
centres_lat = (LAT_1 + LAT_2 + LAT_3) / 3

carte = folium.Map(location=[46.5, 2.5], zoom_start=6)

# Ajouter des marqueurs pour chaque centre de bin
for lon, lat in zip(centres_lon, centres_lat):
    folium.Marker([lat, lon], icon=folium.Icon(color='blue')).add_to(carte)

carte
```
